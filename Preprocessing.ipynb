```python
import pandas as pd
```

```python
crop_data = pd.read_csv('Data-raw/cpdata.csv')
crop_data
```

```python
crop_data.nunique()
```

```python
def low(x):  #converting values to lower case
    return str(x).lower()

crop_data['label'] = crop_data['label'].apply(low)
```

```python
crop_names = crop_data.label.unique()
crop_names
```

```python
fert = pd.read_csv('Data-raw/Fertilizer.csv')
fert
```

```python
fert['Crop'] = fert['Crop'].apply(low) #applying lower case func
```

```python
fert.Crop.unique()
```

```python
empty = []                       #func for listing the matching values(crop names) from fert data
data_in_crops = []
for crop in crop_names:
    empty.append(fert[fert["Crop"]==crop])
    if len(fert[fert["Crop"]==crop]) > 0:
        data_in_crops.append(fert[fert["Crop"]==crop])
```

```python
data_in_crops
```

```python
len(data_in_crops)
```

```python
final_crop = pd.DataFrame()
```

```python
for i in range(0,15):                   
    var = pd.concat([final_crop,data_in_crops[i]])
    final_crop = var
```

```python
final_crop
```

```python
final_crop.rename(columns={"Crop":"label"}, inplace=True)
```

```python
final_crop
```

```python
grouped_data = final_crop.groupby(by="label")["N","P","K","pH"].mean()
grouped_data
```

```python
final_data = pd.merge(crop_data,grouped_data,on="label")
final_data.drop(columns="pH", inplace=True)
final_data
```

```python
final_data["label"].unique()
```

```python
final_data.to_csv("Untitled Folder/Final_data.csv")
```
