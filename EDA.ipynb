```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
final_data = pd.read_csv("Final_data.csv")
final_data
```

```python
final_data.drop("Unnamed: 0", axis=1, inplace=True)
final_data.rename(columns={"label":"crop_name"}, inplace=True)
```

```python
final_data
```

```python
plt.figure(figsize=(20,20))
sns.countplot(final_data.crop_name)
```

```python
col = final_data.columns
for data in col:
    plt.plot(final_data[data])
    plt.ylabel(data)
    plt.show()
```

```python
cropname = final_data.crop_name.unique()
cropname
```

```python
for each_col in col:
    for cn in cropname:
        plt.plot(final_data[each_col][final_data["crop_name"]==cn])
        plt.xlabel(cn)
        plt.ylabel(each_col)
        plt.show()
```
